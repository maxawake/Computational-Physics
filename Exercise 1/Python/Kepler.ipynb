```python
#we find m1=m2=1 ->my=m1*m2/(m1+m2)=1/2
# stable circular orbit vel. after Kepler: v0=sqrt(my/r) with r=0,my=1/2->v0=1/sqrt(2)
#dt: h=dt/T0, T0=r/v0=1/sqrt(2)->h=dt/sqrt(2)







#Orbit with R=1, v0=1

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


#Initial values


x0=1
y0=0
vx0=0
vy0=1


#Integration values    

N=int(1e5)     #timesteps
h=0.01/np.sqrt(2)         #dt

#Position and velocity array
W=np.zeros((N+1,2))
S=np.zeros((N+1,2))

W[0]=[x0,y0]
S[0]=[vx0,vy0]        #initializing each



#Euler step by step integration
for i in range(N):
    
    
    
    S[i+1]=S[i]+h*W[i]                         #New pos. from previous pos. and vel.
    F=-S[i+1]/np.linalg.norm(S[i+1])**3        #Gravitational Force for recent pos. (=acc.)
    W[i+1]=W[i]+h*F                            #Vel. with new acc.
    
  
    x=S[:,0]        
    y=S[:,1]                                    #Slicing variables put of array

    
#Plot format    
plt.figure(figsize=[8,8])

plt.xlabel('x Koordinate' " " r'$\frac{m}{R_0}$', size=10)
plt.ylabel('y Koordinate' " " r'$\frac{m}{R_0}$', size=10)
plt.title('Simulation of 2 body problem with Euler step integration', size=13)


#plot data
plt.plot(x,y, linestyle='-')
```
