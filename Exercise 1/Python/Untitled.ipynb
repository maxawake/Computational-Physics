```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#Function
def f(a,n_0,n_1,y_0):
    #Initializing
    y_n=y_0
    
    #Calculating iteration variables
    O = np.minimum(n_0,n_1)
    A = np.maximum(n_0,n_1)
    n_t = A - O
    Y = np.zeros(n_t)
    
    #Iteration
    if (n_0 < n_1):
        #Forward iteration
        for i in range(O,A,1):
            if i==0: y_n=y_0
            elif i==O: y_n=y_0
            else:
                y_n=(1/i-a*y_n)
                print("Forward:",i,y_n)
                Y[i-A]=y_n
    else:
        #Backward iteration
        for i in range(A,O,-1):
            if i==0: y_n=y_0
            else:
                y_n=(1/i-a*y_n)
                print("Backward:",i,y_n)
                Y[i-A]=y_n
    #Plot
    x=np.arange(O,A,1)        
    plt.plot(x,Y)
    plt.title("Forward iteration")
```

```python
#Values
a = 5
n_0 = 0
n_1 = 30
y_0 = np.log((1+a)/a)

f(a,n_0,n_1,y_0)
```
