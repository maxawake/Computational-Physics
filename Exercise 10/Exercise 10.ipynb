```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random(a,m,c,n,I_0):
    I = np.zeros(n)
    for i in range(n):
        if i==0: I[0]=I_0
        else:
            I[i] = ( a * I[i-1] + c ) % m
    return I

def normal(i,I_0):
    a = 106
    m = 6076
    c = 1283
    r = random(a,m,c,i,I_0)/(m-1)
    return r
```

```python
def f(x):
    return x**2
```

```python
lin = np.linspace(0,1,100)
plt.plot(lin,f(lin))
```

```python
N = 10000
r = np.random.rand(N)
norm = normal(N,0)
func = np.zeros(N)

for i in range(N-1):
    r[i] = norm[i]

for i in range(N-1):
    func[i] = f(r[i])
    
plt.bar(r,func,width=0.005)
plt.plot(lin,f(lin), color="orange")

I = np.sum(func)/N
print("Integral from 0 to 1:",I)
```

```python
def minmax(test, *args):
    res = args[0]
    for arg in args[1:]:
        if test(arg, res):
            res = arg
    return res

def lessthan(x, y): return x < y     
def grtrthan(x, y): return x > y

print(minmax(lessthan, 4, 2, 1, 5, 6, 3))
print(minmax(grtrthan, 4, 2, 1, 5, 6, 3))
```
