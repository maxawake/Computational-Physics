```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random(a,m,c,n):
    I = np.zeros(n)
    for i in range(n):
        I[i] = ( a * I[i-1] + c ) % m
    return I
```

```python
a = 106
m = 6076
c = 1283
random(a,m,c,6)
```

```python
def normal(a,m,c,i):
    r = random(a,m,c,i)/(m-1)
    return r
```

```python
print(normal(106,6076,1283,7))
```

```python
n = 5000
x = normal(a,m,c,n)
y = normal(a,m,c,n+1)


plt.scatter(x,y[1:])
```

```python
def dice(n):
    norm = normal(a,m,c,n)
    dice = np.zeros(n)
    for i in range(n):
        dice[i] = int(norm[i]*6+1)
      #  print(dice[i])
    return dice
```

```python
dice(10)
```

```python
n = 100000
y = np.zeros(n*10)
y = dice(n*10)
hist = np.zeros(n)
print(len(y))
for i in range(n):
    hist[i] = np.sum(y[i*10:(10*i+11)])
  #  print(i, y[i*10:(10*i+11)], hist[i])
#print(hist)
plt.hist(hist,bins=60, width=1)
```
