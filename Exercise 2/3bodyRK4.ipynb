```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

G=1
m1=1
m2=1
m3=1

#Set initial positions and velocities for m1, m2 and m3

R1=(-0.97000436, 0.24308753)
R2=(-0.46620368, -0.43236573)
R3=(0.9700436, -0.24308753)

V1=(-0.46620368, -0.43236573)
V2=(0.0, 0.0)
V3=(0.93240737, 0.86473146)

n=1e5      #Integrationsteps
t=0       #Start time
te=10     #End time

h=te/n     #time step size



def f(Y,t):
    
    #storing positin and velocity for masses m1, m2 and m3
    
    (x1,y1)=Y[0]
    (x2,y2)=Y[1]
    (x3,y3)=Y[2]
    
    (vx1,vy1)=Y[3]
    (vx2,vy2)=Y[4]
    (vx3,vy3)=Y[5]
    
    a1=m2*((Y[1,]-Y[0,])/(np.linalg.norm(Y[1,]-Y[0,])**3))+m3*((Y[2,]-Y[0,])/(np.linalg.norm(Y[0,]-Y[2,])**3))    #accelaration of m1
    a2=m3*((Y[2,]-Y[1,])/(np.linalg.norm(Y[2,]-Y[1,])**3))+m1*((Y[0,]-Y[1,])/(np.linalg.norm(Y[1,]-Y[0,])**3))    #accelaration of m2
    a3=m1*((Y[0,]-Y[2,])/(np.linalg.norm(Y[0,]-Y[2,])**3))+m2*((Y[1,]-Y[2,])/(np.linalg.norm(Y[2,]-Y[1,])**3))    #accelaration of m2
    
    return np.array([Y[3], a1,Y[4], a2,Y[5], a3])

Y=np.array([R1,R2,R3,V1,V2,V3])        #set array with initial values

x1=[Y[0,0]]
y1=[Y[0,1]]
vx1=[Y[3,0]]
vy1=[Y[3,1]]
x2=[Y[1,0]]
y2=[Y[1,1]]
vx2=[Y[4,0]]
vy2=[Y[4,1]]
x3=[Y[2,0]]
y3=[Y[2,1]]
vx3=[Y[5,0]]
vy3=[Y[5,1]]


while t<te:
    
    k1=h*f(Y,t)
    k2=h*f(Y+0.5*k1,t+h/2)
    k3=h*f(Y+0.5*k2,t+h/2)
    k4=h*f(Y+k3,t+h)
    
    Y+=(k1+2*k2+2*k3+k4)/6
    
    x1.append(Y[0,0])
    y1.append(Y[0,1])
    vx1.append(Y[3,0])
    vy1.append(Y[3,1])
    x2.append(Y[1,0])
    y2.append(Y[1,1])
    vx2.append(Y[4,0])
    vy2.append(Y[4,1])
    x3.append(Y[2,0])
    y3.append(Y[2,1])
    vx3.append(Y[5,0])
    vy3.append(Y[5,1]) 
    
    t += h
    
    
plt.plot(x1,y1, label='m1')
plt.plot(x2,y2, label='m2') 
plt.plot(x3,y3, label='m3')
#plt.xscale('log')
#plt.yscale('log')
plt.legend()
plt.show()
    
```
